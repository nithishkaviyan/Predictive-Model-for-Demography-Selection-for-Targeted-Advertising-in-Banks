```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import plotly.offline as py
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected=True)

import plotly.graph_objs as go
```

```python
##Read data
data=pd.read_csv("bankadditionalfull.csv",sep=';')
data
```

```python
##Dependant variable Y
trace1=go.Bar(
   x=list(data.y.unique()),
   y=[data[data.y=='yes'].y.count()],
   name='YES')

trace2=go.Bar(
   x=list(data.y.unique()),
   y=[data[data.y=='no'].y.count()],
   name='NO')

temp_dat=[trace1,trace2]

layout=go.Layout(title="NUMBER OF INSTANCES OF DIFFERENT CATEGORIES OF 'Y'",barmode='group',
                 xaxis=dict(title='Y',
                            titlefont=dict(family='Calibri',size=16)),
                 yaxis=dict(title='Count',
                            titlefont=dict(family='Calibri',size=16))
                )
fig=go.Figure(data=temp_dat,layout=layout)

py.iplot(fig, filename='grouped-bar')
```
