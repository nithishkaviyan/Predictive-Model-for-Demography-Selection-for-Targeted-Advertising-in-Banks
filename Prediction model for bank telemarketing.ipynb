```python
##Changing values of y to 0 and 1
data_copy=data.copy()
data_copy.y[data_copy.y=='yes']=1
data_copy.y[data_copy.y=='no']=0
data_copy.y=data_copy.y.astype('int64')
data_copy=pd.get_dummies(data_copy)
```

```python
##Shifting values of y to last column and random shuffling
data_arr=np.array(data_copy)
temp_arr=np.reshape(data_arr[:,10],(-1,1))
data_arr=np.delete(data_arr,10,1)
```

```python
##Scale data
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(data_arr[:,:-1])
data_arr=scaler.transform(data_arr[:,:-1])
```

```python
##Principal Component Analysis
from sklearn.decomposition import PCA
pca=PCA(n_components=5)
data_arr=pca.fit_transform(data_arr[:,:-1])
per_var=pca.explained_variance_ratio_
np.round((per_var*100.0),2)
```

```python
print("Percentage of variance explained by Principal Components: "+str(np.round(per_var.sum()*100.0,2))+"%")
```

```python
##PCA plot
color=['blue','green']
plt.scatter(data_arr[:,0],data_arr[:,1],color=color)
plt.show()
```

From the PCA plots, it can be seen that the components aren't scattered in random. This is due to the presence of large number of categorical variables which has some higher order of correlation than just linear relationship. Hence, PCA is not an appropriate metric in this case. Also, reducing a large number of variables (63) to 5 principal components may incur loss of information from the data. Hence the predictive model were developed using the original data, not the principal components.    

```python
##Split data to train and test set by 80:20
data_arr=np.hstack((data_arr,temp_arr))
np.random.shuffle(data_arr)
tr_set=int(0.8*len(data_arr))
x_train=data_arr[0:tr_set,:-1]
y_train=data_arr[0:tr_set,-1]
x_test=data_arr[tr_set:,:-1]
y_test=data_arr[tr_set:,-1]
```

## k-Nearest Neighbor Algorithm

```python
##k-Nearest Neighbor
from sklearn.neighbors import KNeighborsClassifier
knn_acc=[]
knn_model=KNeighborsClassifier(n_neighbors=9)
knn_model.fit(x_train,y_train)
knn_predict=knn_model.predict(x_test)
knn_score=knn_model.score(x_test,y_test)
knn_acc.append(knn_score)
print("Accuracy of the model (k="+str(9)+"): "+str(knn_score))
```

```python
##But the given data has more examples of class 'no' than 'yes'. Hence accuracy is not an appropriate measure
##Confusion Matrix would be an appropriate measure
from sklearn.metrics import confusion_matrix
conf_mat=confusion_matrix(y_test,knn_predict)
```

```python
tn=conf_mat[0,0]
fn=conf_mat[0,1]
tp=conf_mat[1,1]
fp=conf_mat[1,0]
print("True Positive: "+str(tp))
print("True Negative: "+str(tn))
print("False Positive: "+str(fp))
print("False Negative: "+str(fn))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,knn_predict,target_names=['Class NO', 'Class YES']))
```

```python
##Class YES has a very low f-1 score. Hence we may conclude that the model doesn't fit well to the data.  
#However class YES has very little instances in the data. Hence, a stratified k-fold class validation would be another good approach to estimate the performance of the model.
from sklearn.model_selection import cross_val_score
cv_precision=cross_val_score(knn_model,data_arr[:,:-1],data_arr[:,-1],cv=5,scoring='precision')
cv_recall=cross_val_score(knn_model,data_arr[:,:-1],data_arr[:,-1],cv=5,scoring='recall')
cv_accuracy=cross_val_score(knn_model,data_arr[:,:-1],data_arr[:,-1],cv=5,scoring='accuracy')
cv_f1=cross_val_score(knn_model,data_arr[:,:-1],data_arr[:,-1],cv=5,scoring='f1')
```

```python
print("Results for 5 fold cross validation")
print("Mean Precision: "+str(np.round(np.mean(cv_precision),4))+" (+/-"+str(np.round(abs(cv_precision.std()),4))+")")
print("Mean Recall: "+str(np.round(np.mean(cv_recall),4))+" (+/-"+str(np.round(abs(cv_recall.std()),4))+")")
print("Mean Accuracy: "+str(np.round(np.mean(cv_accuracy),4))+" (+/-"+str(np.round(abs(cv_accuracy.std()),4))+")")
print("Mean f1-score: "+str(np.round(np.mean(cv_f1),4))+" (+/-"+str(np.round(abs(cv_f1.std()),4))+")")
```
